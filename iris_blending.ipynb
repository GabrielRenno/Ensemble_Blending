### Import the dataset

```python
# Import iris dataset
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
# Import pandas
import pandas as pd

# Create features and target
X = iris.data
y = iris.target

# Create dataframe
df = pd.DataFrame(X, columns=iris.feature_names)

# Append target and class to dataframe
df['target'] = y
df
```

### Normalize the data

```python
# Normalize data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_norm = scaler.fit_transform(X)
```

```python
# Plt values counts of target
import matplotlib.pyplot as plt
plt.hist(y)
```

```python
features_normalized =X_norm
target = y
```

### Train test split

```python
# Train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features_normalized, target, test_size=0.2, random_state=42)

# Hold out set
X_train, X_holdout, y_train, y_holdout = train_test_split(X_train, y_train, test_size=0.25, random_state=42)
```

```python
# Print the name and the len for each set
for name, set in zip(['Train', 'Holdout', 'Test'], [X_train, X_holdout, X_test]):
    print(name, len(set))
```

### Train the first layer models using the training data

```python
# Import classification models
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

# Add decision tree
model_1 = DecisionTreeClassifier()
model_1.fit(X_train, y_train)
holdout_pred_1=model_1.predict(X_holdout)
test_pred_1=model_1.predict(X_test)
holdout_pred_1=pd.Series(holdout_pred_1)

# Add KNN
model_2 = KNeighborsClassifier()
model_2.fit(X_train, y_train)
holdout_pred_2=model_2.predict(X_holdout)
test_pred_2=model_2.predict(X_test)
holdout_pred_2=pd.Series(holdout_pred_2)


# Add naive bayes
from sklearn.naive_bayes import GaussianNB
model_3 = GaussianNB()
model_3.fit(X_train, y_train)
holdout_pred_3=model_3.predict(X_holdout)
holdout_pred_3=pd.Series(holdout_pred_3)

test_pred_3=model_3.predict(X_test)

# Add SVM
from sklearn.svm import SVC
model_4 = SVC()
model_4.fit(X_train, y_train)
holdout_pred_4=model_4.predict(X_holdout)
holdout_pred_4=pd.Series(holdout_pred_4)

test_pred_4=model_4.predict(X_test)

# Add logistic regression
model_5 = LogisticRegression()
model_5.fit(X_train, y_train)
holdout_pred_5=model_5.predict(X_holdout)
holdout_pred_5=pd.Series(holdout_pred_5)

test_pred_5=model_5.predict(X_test)

# Add random forest
from sklearn.ensemble import RandomForestClassifier
model_6 = RandomForestClassifier()
model_6.fit(X_train, y_train)
holdout_pred_6=model_6.predict(X_holdout)
holdout_pred_6=pd.Series(holdout_pred_6)

test_pred_6=model_6.predict(X_test)


# Add XGBoost
from xgboost import XGBClassifier
model_7 = XGBClassifier()
model_7.fit(X_train, y_train)
holdout_pred_7=model_7.predict(X_holdout)
holdout_pred_7=pd.Series(holdout_pred_7)

test_pred_7=model_7.predict(X_test)



```

```python

# Create a dataframe with accuracy on train and holdout
models = [model_1, model_2, model_3, model_4, model_5, model_6, model_7]
model_names = ['Decision Tree', 'KNN', 'Naive Bayes', 'SVM', 'Logistic Regression', 'Random Forest', 'XGBoost']
train_accuracy = []
holdout_accuracy = []
test_accuracy = []
for model in models:
    train_accuracy.append(model.score(X_train, y_train))
    holdout_accuracy.append(model.score(X_holdout, y_holdout))
    test_accuracy.append(model.score(X_test, y_test))

# Create the dataframe
df = pd.DataFrame({'Model': model_names, 'Train Accuracy': train_accuracy, 'Holdout Accuracy': holdout_accuracy, 'Test Accuracy': test_accuracy})
df

# Plot a bar chart of df
df.plot.bar(x='Model', y=['Train Accuracy', 'Holdout Accuracy', 'Test Accuracy'], rot=45, figsize=(10, 6))






```

```python
# Create a dataframe with the hopldouts predictions and name the columns by the model names and the y_holdout
df = pd.DataFrame({'model_1': holdout_pred_1, 'model_2': holdout_pred_2,'model_3': holdout_pred_3,
                     'model_4': holdout_pred_4,'model_5': holdout_pred_5,'model_6': holdout_pred_6,
 'y_holdout': y_holdout})
```

```python
df
```

### Train the meta learner using the predictions of the first layer models

```python
# Run a logistic regression on the holdout predictions
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(df.drop('y_holdout', axis=1), df['y_holdout'])

# Predict
pred = model.predict(df.drop('y_holdout', axis=1))

# Print the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(df['y_holdout'], pred))


```

### Test

```python
# Make predictions in the first layer
pred_1 = model_1.predict(X_test)
pred_2 = model_2.predict(X_test)
pred_3 = model_3.predict(X_test)
pred_4 = model_4.predict(X_test)
pred_5 = model_5.predict(X_test)
pred_6 = model_6.predict(X_test)
pred_7 = model_7.predict(X_test)

# Create a dataframe with the predictions and name the columns by the model names and the y_test
df = pd.DataFrame({'model_1': pred_1, 'model_2': pred_2,'model_3': pred_3,
                        'model_4': pred_4,'model_5': pred_5,'model_6': pred_6,
    'y_test': y_test})
                        
```

```python
df.head(3)
```

```python
# Make predictions in the second layer
pred = model.predict(df.drop('y_test', axis=1))

# Print the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(df['y_test'], pred))

# Dataframe with the predictions
df = pd.DataFrame({'model_1': pred_1, 'model_2': pred_2,'model_3': pred_3,
                        'model_4': pred_4,'model_5': pred_5,'model_6': pred_6,'meta_model': pred,
    'y_test': y_test})
```

```python
df
```

### Evaluate the models

```python
# Confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(df['y_test'], df['meta_model'])

# Print confusion matrix
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(confusion_matrix(df['y_test'], df['meta_model']), annot=True, fmt='d')
plt.show()
```

```python
# Plot classification report
from sklearn.metrics import classification_report
print(classification_report(df['y_test'], df['meta_model']))
```
